```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('X_train.csv')
target_data = pd.read_csv('targets_train.csv')
```

```python
# creating datasets

new_data = {}
for col in target_data.columns:
    data = pd.DataFrame()
    for columns in df.columns:
        if columns.startswith('Q') or columns.startswith('Y'):
            if col[-2:] == columns[-2:]:
                data[columns] = df[columns]
        data[col] = target_data[col]
        new_data[col] = data
```

```python
target_4 = new_data['target_4']
```

```python
target_4
```

```python
import matplotlib.pyplot as plt
def distplot(rows,col,data,feature):
    """Returns the distribution plots."""
    fig,ax = plt.subplots(nrows=rows,ncols=col,sharey=True,figsize=(10,10))
    fig.suptitle("Distribution of data.")


    feature_data = data.drop([feature],axis=1)
    axes = ax.flatten()  # flatten the 2D array of axes into a 1D array
    for i, col in enumerate(feature_data,start=-1):
        axes[i].scatter(feature_data[col], data[feature])
        axes[i].set_title(col,fontsize=8)  # set the title of each subplot to the column name
        axes[i].grid(True)
    plt.show()
    plt.tight_layout()
```

```python
distplot(4,2,target_4,'target_4')
```

```python
target_4['target_4_cumsum'] = target_4['target_4'].cumsum()
target_4.head()
```

```python
for col in target_4.drop(['target_4','target_4_cumsum'],axis=1):
    target_4[col] = target_4[col].cumsum()
```

```python
selected_data = target_4.drop('target_4',axis=1)
distplot(4,2,target_4,'target_4_cumsum')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import r2_score
from sklearn.linear_model import Lasso


X_4 = target_4.drop(['target_4','target_4_cumsum'],axis=1)
Y_4 = target_4['target_4_cumsum']
xtrain_4,xtest_4,ytrain_4,ytest_4 = train_test_split(X_4,Y_4,test_size=0.25,random_state=3)
pls_reg_4 = PLSRegression(n_components=4)
pls_reg_4.fit(xtrain_4,ytrain_4)

y_pred_train = pls_reg_4.predict(xtrain_4)
y_pred_test = pls_reg_4.predict(xtest_4)


xtrain_trans = pls_reg_4.transform(xtrain_4)
xtest_trans = pls_reg_4.transform(xtest_4)


lasso = Lasso(alpha=0.01)
lasso.fit(xtrain_4,ytrain_4)

y_lasso_train = lasso.predict(xtrain_4)
y_lasso_test = lasso.predict(xtest_4)

print(f"r2_score(lasso) on train is {r2_score(ytrain_4,y_lasso_train)} and test is {r2_score(ytest_4,y_lasso_test)}")
print(f"r2 score(PLS) on train set is {r2_score(ytrain_4,y_pred_train)} and test set is {r2_score(ytest_4,y_pred_test)}")
```

```python
from sklearn.metrics import mean_squared_error

print(f"mse score on train set is {mean_squared_error(y_pred_train,ytrain_4)} and test set is {mean_squared_error(y_pred_test,ytest_4)}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

rdclf = RandomForestRegressor(max_depth = 15, min_samples_split=3, n_estimators=150)
rdclf.fit(xtrain_trans,ytrain_4)

ypred_rf_train = rdclf.predict(xtrain_trans)
ypred_rf_test = rdclf.predict(xtest_trans)

print(f"r2 score on {r2_score(ytrain_4,ypred_rf_train)} and test set is {r2_score(ytest_4,ypred_rf_test)}.")
print(f"MSE score on {mean_squared_error(ytrain_4,ypred_rf_train)} and test set is {mean_squared_error(ytest_4,ypred_rf_test)}.")
```
